```python
class Marsupial:
    def __init__(self):
        self.itmLst = []
    
    def put_in_pouch(self,itm):
        self.itmLst.append(itm)
        
    def pouch_contents(self):
        return self.itmLst

class Kangaroo(Marsupial):
    def __init__(self, dx, dy):
        self.x = dx
        self.y = dy
        
    def jump(self,dx,dy):
        self.x +=dx
        self.y +=dy
        
    def __str__(self):
        return "I am a Kangaroo located at coordinates ({0},{1})".format(self.x,self.y)
```

```python
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()
```

```python
## Reference https://docs.python.org/3/library/html.parser.html
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    (h1,h2,h3,h4,h5,h6) = (False, False,False,False,False,False)
    
    def handle_starttag(self, tag, attrs):
        if tag == 'h1':
            self.h1 = True
        if tag == 'h2':
            self.h2 = True
        if tag == 'h3':
            self.h3 = True
        if tag == 'h4':
            self.h4 = True
        if tag == 'h5':
            self.h5 = True
        if tag == 'h6':
            self.h6 = True
            

    def handle_endtag(self, tag):
        if tag == 'h1':
            self.h1 = False
        if tag == 'h2':
            self.h2 = False
        if tag == 'h3':
            self.h3 = False
        if tag == 'h4':
            self.h4 = False
        if tag == 'h5':
            self.h5 = False
        if tag == 'h6':
            self.h6 = False
            

    def handle_data(self, data):
        if self.h1:
            print(data)
        elif self.h2:
            print("\t{}".format(data))
        elif self.h3:
            print("\t\t{}".format(data))
        elif self.h4:
            print("\t\t\t{}".format(data))
        elif self.h5:
            print("\t\t\t\t{}".format(data))
        elif self.h6:
            print("\t\t\t\t\t{}".format(data))
        
        
inputfile = open('w3c.html')
file_content = inputfile.read()
inputfile.close()
headingFormatParser = HeadingParser()

#HTML Content
# file_content = '<!DOCTYPE html> <html lang="en"> <head> <meta name="description" content="Webpage description goes here" /> <meta charset="utf-8"> <title>Change_me</title> <meta name="viewport" content="width=device-width, initial-scale=1"> <meta name="author" content=""> <link rel="stylesheet" href="css/style.css"> </head> <body> <h1>W3C Mission (H1)</h1> <h2>Principles (H2)</h2> <h3>This is H3 (H3)</h3> <h4>This is H4 (H4)</h4> </body> </html>'
headingFormatParser.feed(file_content)
```

```python
from urllib.parse import urljoin
from html.parser import HTMLParser
from urllib.request import urlopen
class Collector(HTMLParser):
    'collects hyperlink URLs into a list'

    def __init__(self, url):
        'initializes parser, the url, and a list'
        HTMLParser.__init__(self)
        self.url = url
        self.links = []

    
    def handle_starttag(self, tag, attrs):
        'collects hyperlink URLs in their absolute format'
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    # construct absolute URL
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http': # collect HTTP URLs
                        self.links.append(absolute)
                        
    def getLinks(self):
        'returns hyperlinks URLs in their absolute format'
        return self.links

```

```python
def webdir(url,depth,indent):
    
    if(depth==0):
        return
    resource = urlopen(url)
    content = resource.read().decode()
    collector = Collector(url)
    collector.feed(content)

    for url in collector.getLinks()[0:depth+1]:
        print("\t"*indent +url)
        webdir(url,depth-1,indent+1)
    
```

```python
url = 'https://www.w3schools.com/python/default.asp'
webdir(url,3,0)
```
